```python
Long_latit = pd.read_csv('https://raw.githubusercontent.com/warazkhan/Data-Analysis/main/countries-latitude-longitude.csv')
Long_latit['name'] = Long_latit['name'].str.upper()
Long_latit.head()
```

<div style="text-align: left;">
    <p>cette partie du code charge des informations sur<b> les latitudes et longitudes</b> des pays dans un DataFrame et normalise les noms des pays en majuscules pour assurer une cohérence dans les données.</p>
    
</div>

```python
df['Country'] = df['Country'].str.strip()
Long_latit['name'] = Long_latit['name'].str.strip()

df['longitude'] = df['Country'].map(Long_latit.set_index('name')['longitude'])
df['latitude'] = df['Country'].map(Long_latit.set_index('name')['latitude'])
df.head()
```

<div style="text-align: left;">
    <p>ces lignes de code mettent à jour le DataFrame df en ajoutant les colonnes <b>'longitude'</b> et <b>'latitude'</b>, qui sont obtenues en associant les noms de pays de la colonne <b>'Country'</b> avec les données de latitude et longitude du DataFrame <b>Long_latit</b></p>
    
</div>

```python
df_map = df.copy()
```

```python
df_map.dropna(subset = ['longitude'] , axis = 0 , inplace = True)
```

```python
df_map.head()
```

```python
df_map.shape
```

```python
fatalities_sum = df_map.groupby('Country')['Fatalities'].sum().reset_index()


world_map = folium.Map(
    location=[48.8566, 2.3522],
    tiles='OpenStreetMap',
    zoom_start=2
    )

for index, row in df_map[['Country', 'latitude', 'longitude']].iterrows():
    country_info = fatalities_sum[fatalities_sum['Country'] == row['Country']]
    fatalities = country_info['Fatalities'].values[0] if not country_info.empty else 0

    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"{row['Country']} - Fatalities: {fatalities}"
    ).add_to(world_map)


world_map.save("D:\Competition\pic\world_map.html")


display(world_map)


```
